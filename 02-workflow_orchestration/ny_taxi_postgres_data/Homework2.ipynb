```python
# 02-workflow_orchestration/ny_taxi_postgres_data/Homework2.ipynb
import pandas as pd

# Define the combinations
combinations = [
    ("yellow", 2021, 1),
    ("yellow", 2021, 2),
    ("yellow", 2021, 3),
    ("green", 2021, 1),
    ("green", 2021, 2),
    ("green", 2021, 3),
]

for taxi, year, month in combinations:
    # Verifica si el taxi es "yellow" o "green"
    if taxi == "yellow":
        url = f"https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow/{taxi}_tripdata_{year}-{str(month).zfill(2)}.csv.gz"
    else:
        url = f"https://github.com/DataTalksClub/nyc-tlc-data/releases/download/green/{taxi}_tripdata_{year}-{str(month).zfill(2)}.csv.gz"

    try:
        df = pd.read_csv(url, compression='gzip', low_memory=False)
        rows = len(df)
        print(f"Taxi {taxi}, Año {year}, Mes {month}: {rows} filas")
    except Exception as e:
        print(f"Error en taxi {taxi}, año {year}, mes {month}: {e}")
```

```python
# DESCARGAR 2020 Y 2021 Y RESPONDER PREGUNTAS 
import pandas as pd

def get_row_count(taxi, year, month):
    # Formatear el mes a dos dígitos (ej. 3 -> '03')
    month_str = str(month).zfill(2)
    for taxi, year, month in combinations:
        # Verifica si el taxi es "yellow" o "green"
        if taxi == "yellow":
            url = f"https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow/{taxi}_tripdata_{year}-{str(month).zfill(2)}.csv.gz"
        else:
            url = f"https://github.com/DataTalksClub/nyc-tlc-data/releases/download/green/{taxi}_tripdata_{year}-{str(month).zfill(2)}.csv.gz"
    
    print(f"Descargando {taxi} {year}-{month_str}...")
    df = pd.read_csv(url, compression='gzip', low_memory=False)
    return len(df)

# --- PREGUNTA 3 y 4: Totales de 2020 ---
# Nota: Esto puede tardar unos minutos dependiendo de tu conexión
total_yellow_2020 = sum(get_row_count('yellow', 2020, m) for m in range(1, 13))
total_green_2020 = sum(get_row_count('green', 2020, m) for m in range(1, 13))

print(f"Total Yellow 2020: {total_yellow_2020}")
print(f"Total Green 2020: {total_green_2020}")

# --- PREGUNTA 5: Yellow Taxi Marzo 2021 ---
rows_yellow_mar_2021 = get_row_count('yellow', 2021, 3)
print(f"Filas Yellow Marzo 2021: {rows_yellow_mar_2021}")
```
