```python
import pandas as pd
# Define the combinations
combinations = [ ("yellow", 2021, 3), # pregunta 
    ("green", 2021, 1), 
    ("green", 2021, 2), 
    ("green", 2021, 3),]

for taxi, year, month in combinations: # Verifica si el taxi es "yellow" o "green" 
    if taxi == "yellow": url = f"https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow/{taxi}_tripdata_{year}-{str(month).zfill(2)}.csv.gz" 
else: 
    url = f"https://github.com/DataTalksClub/nyc-tlc-data/releases/download/green/{taxi}_tripdata_{year}-{str(month).zfill(2)}.csv.gz" 

try: 
    df = pd.read_csv(url, compression='gzip', low_memory=False) 
    rows = len(df) 
    print(f"Taxi {taxi}, A√±o {year}, Mes {month}: {rows} filas") 

except Exception as e: 
    print(f"Error en taxi {taxi}, a√±o {year}, mes {month}: {e}")
```
